```python
%matplotlib notebook
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

## Daily Birth

### Basic Analysis Rolling moments

```python
series = Series.from_csv('daily-total-female-births.csv', header=0)
plt.figure()
series.plot()
window = 15
rolling_mean = series.rolling(window=window, center=False).mean()
rolling_mean.plot();
rolling_std = series.rolling(window=window, center=False).std()
rolling_std.plot();
rolling_kurtosis = series.rolling(window=window, center=False).kurt()
rolling_kurtosis.plot();
plt.legend(["serie","rolling average", "rolling std", "rolling kurt"]);
```

### ADF

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
from statsmodels.tsa.stattools import adfuller
X = series.values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

### KPSS

```python
from statsmodels.tsa.stattools import kpss
result = kpss(X)
print(result)
```
